# Importing Libraries

```python
import pandas as pd
import numpy as np
from IPython.display import display
import seaborn as sns
import matplotlib.pylab as plt
import warnings
warnings.filterwarnings('ignore')
```

# Reading and copying the dataset

```python
churn = pd.read_csv("Telco-Customer-Churn.csv")
print(churn.shape)
display(churn)
churn_cpy = churn.copy()
```

# Removing Duplicates

```python
churn_cpy = churn_cpy.drop_duplicates()
display(churn_cpy)
```

# Dealing with Nulls

```python
nulls = churn_cpy.isnull().sum().sort_values(ascending=False)
print(nulls)
```

# Features and Target

```python
features_raw = churn_cpy.drop(['customerID'], axis = 1)
features_raw
```

```python
features_raw.info()
```

```python
features_raw['TotalCharges'] = pd.to_numeric(features_raw['TotalCharges'], errors='coerce')
features_raw
```

```python
features_raw.isnull().sum()
```

```python
features_raw = features_raw.dropna()
display(features_raw)
```

```python
features_raw.nunique()
```

```python
features_raw['gender'] = features_raw['gender'].map({'Female': 0, 'Male': 1})
features_raw
```

# Binary Columns

```python
binary_columns = ['SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']

# Convert binary columns to 0 and 1
features_raw[binary_columns] = features_raw[binary_columns].replace({'Yes': 1, 'No': 0})
features_raw
```

```python
features_raw.isnull().sum()
```

# Multicategory columns

```python
# List of columns with more than two unique values
multi_category_columns = ['MultipleLines', 'InternetService', 'Contract',
                          'TechSupport', 'StreamingTV', 'StreamingMovies',
                          'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                          'PaymentMethod']

# Apply one-hot encoding
features = pd.get_dummies(features_raw, columns=multi_category_columns, drop_first=True)  # drop_first=True to avoid dummy variable trap
#print(features.to_string())
display(features)
```

```python
display(features.nunique().to_string())
```

# Converting boolean columns (True & False) into int (0 & 1)

```python
# List of boolean columns to convert
boolean_columns = [
    'MultipleLines_No phone service',
    'MultipleLines_Yes',
    'InternetService_Fiber optic',
    'InternetService_No',
    'Contract_One year',
    'Contract_Two year',
    'TechSupport_No internet service',
    'TechSupport_Yes',
    'StreamingTV_No internet service',
    'StreamingTV_Yes',
    'StreamingMovies_No internet service',
    'StreamingMovies_Yes',
    'OnlineSecurity_No internet service',
    'OnlineSecurity_Yes',
    'OnlineBackup_No internet service',
    'OnlineBackup_Yes',
    'DeviceProtection_No internet service',
    'DeviceProtection_Yes',
    'PaymentMethod_Credit card (automatic)',
    'PaymentMethod_Electronic check',
    'PaymentMethod_Mailed check'
]


# Convert boolean values to 0 and 1
features[boolean_columns] = features[boolean_columns].astype(int)

# Display the updated DataFrame
display(features)
```

```python
features['Churn'] = features['Churn'].map({'No': 0, 'Yes': 1})
features
```

```python
corr = features.corr()['Churn']
corr
```

```python
top_features = corr[(corr > 0.1) | (corr < - 0.1)].sort_values(ascending=False) #displays  correlation coefficient greater than 0.5
top_features = top_features.drop('Churn')
top_features.index
```

```python
target = features['Churn']
target
```

```python
features.drop('Churn' ,axis=1 , inplace=True)
features = features[top_features.index]
features
```

# Splitting the data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
target.value_counts() 
# There is class imbalance in the data
```

# After experimenting many models found best model : Random Forrest

# Random Forrest

### After tuning for long time using GridSearch here is the best model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, accuracy_score
from sklearn.model_selection import ShuffleSplit
from imblearn.over_sampling import SMOTE

# Initialize the Random Forest Classifier with class weights
model = RandomForestClassifier(random_state=0,class_weight='balanced')

# Create the parameter grid
parameters = {
    'n_estimators': [230],          # Number of trees in the forest
    'max_depth': [10],               # Maximum depth of the tree
    'min_samples_split': [2],
    'min_samples_leaf': [2]               # Minimum number of samples required to split an internal node
}

# Set up cross-validation
cv_sets = ShuffleSplit(n_splits=5, random_state=0, test_size=0.2, train_size=None)

# Perform grid search on the classifier using F1 score as the scoring method
grid_rf = GridSearchCV(model, parameters, scoring='f1', cv=cv_sets)

# Fit the grid search object to the training data and find the optimal parameters
grid_rf.fit(X_train, y_train)

# Get the best estimator
best_rf = grid_rf.best_estimator_

# Fit the new model
best_rf.fit(X_train, y_train)

# Make predictions using the new model
best_train_predictions_rf = best_rf.predict(X_train)
best_test_predictions_rf = best_rf.predict(X_test)

# Calculate and print the F1 score and accuracy
print('Random Forest Training F1 Score:', f1_score(best_train_predictions_rf, y_train))
print('Random Forest Testing F1 Score:', f1_score(best_test_predictions_rf, y_test))
print('Random Forest Training Accuracy:', accuracy_score(y_train, best_train_predictions_rf))
print('Random Forest Testing Accuracy:', accuracy_score(y_test, best_test_predictions_rf))

# Explore what parameters ended up being used in the new model
print('Best Parameters:', grid_rf.best_params_)
```

```python
# Function to check the output for a specific customer using Random Forest
def check_customer_output_by_index(customer_index, model, X_test, y_test=None):
    
        # Retrieve the customer's data by index
    customer_data = X_test.iloc[customer_index]

        # Reshape the data to be 2D for prediction
    customer_data = customer_data.values.reshape(1, -1)

        # Predict the outcome for the customer
    customer_prediction = best_rf.predict(customer_data)
    customer_proba = best_rf.predict_proba(customer_data)

        # Output the result
    print(f"Prediction for Customer at index {customer_index}: {customer_prediction[0]}")

        # If true labels exist, compare the prediction to the actual label
    if y_test is not None:
        actual_output = y_test.iloc[customer_index]
        print(f"Actual Output for Customer at index {customer_index}: {actual_output}")
    
    return customer_prediction , actual_output

# Example usage:
# Replace 'customer_index' with the row index of the customer in your test dataset

# Replace with the actual index of the customer you want to check
right , wrong = 0 , 0

for i in range(0,30):
    customer_index = i * 10
    customer_prediction , actual_output = check_customer_output_by_index(customer_index, model, X_test, y_test)
    if customer_prediction == actual_output :
        right = right + 1
    else :
        wrong = wrong + 1 

print()
print("Right = ",right)
print()
print("Wrong = ",wrong)
```
